```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Données
X = np.array([12000, 15000, 20000, 25000, 30000]).reshape(-1, 1)
y = np.array([50000, 70000, 100000, 140000, 180000])
# Modèle
model = LinearRegression()
model.fit(X, y)
# Prédiction
revenu_test = np.array([[2200]])
montant_pred = model.predict(revenu_test)
print("Montant du prêt estimé :", montant_pred)
# Visualisation
plt.scatter(X, y, color='blue', label="Données réelles")
plt.plot(X, model.predict(X), color='red', label="Régression linéaire")
plt.xlabel("Revenu mensuel (€)")
plt.ylabel("Montant du prêt (€)")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
# Données
X = np.array([10000, 13000, 16000, 20000, 23000, 28000]).reshape(-1, 1)
y = np.array([0, 0, 0, 1, 1, 1])
# Création du modèle de régression logistique
model = LogisticRegression()
model.fit(X, y)
# Prédiction pour un nouveau client
revenu_test = np.array([[18000]])
prob_sol = model.predict_proba(revenu_test)[0, 1]  # probabilité d'être solvable
prediction = model.predict(revenu_test)[0]
# classification (0 ou 1)
print(f"Pour un revenu de {revenu_test[0,0]} Dh :")
print(f"Probabilité d'être solvable = {prob_sol:.2f}")
print(f"Prédiction du modèle = {'Solvable' if prediction==1 else 'Non solvable'}")
# Visualisation
plt.scatter(X, y, color='blue', label="Données réelles")
X_test = np.linspace(9000, 30000, 500).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]
plt.plot(X_test, y_prob, color='red', label="Probabilité prédite")
plt.xlabel("Revenu mensuel (Dh)")
plt.ylabel("Probabilité de solvabilité")
plt.legend()
plt.show()
```

```python
# Importation des bibliothèques
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
# Données
X = np.array([10000, 13000, 16000, 20000, 23000, 28000]).reshape(-1, 1)
y = np.array([0, 0, 0, 1, 1, 1])
# Création et entraînement du modèle
model = LogisticRegression()
model.fit(X, y)
# Prédictions
y_pred = model.predict(X)
# Calcul de la matrice de confusion
cm = confusion_matrix(y, y_pred)
print("Matrice de confusion :\n", cm)
```

```python
import csv
import matplotlib.pyplot as plt
chemin = r"D:\TP\meteo_maroc.csv"

dates = []
villes = []
temperatures = []

with open(chemin, mode='r', encoding='utf-8') as fichier:
    lecteur = csv.DictReader(fichier)
    for ligne in lecteur:
        dates.append(ligne['Date'])
        villes.append(ligne['Ville'])
        temperatures.append(float(ligne['Température (°C)']))
villes_uniques = []

for ville in villes:
    if ville not in villes_uniques:
        villes_uniques.append(ville)

for ville in villes_uniques:
    temp_ville = []
    dates_ville = []

    for i in range(len(villes)):
        if villes[i] == ville:
            temp_ville.append(temperatures[i])
            dates_ville.append(dates[i])

    plt.figure()
plt.plot(temperatures, marker='o')
plt.title("Évolution globale de la température")
plt.xlabel("Index des observations")
plt.ylabel("Température (°C)")
plt.grid()
plt.show()
```
