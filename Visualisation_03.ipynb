```python
#Étape 1 : Import des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Étape 2 : Génération de la série temporelle
import numpy as np
import pandas as pd

np.random.seed(10)  # reproductibilité

# Création des dates mensuelles sur 3 ans
dates = pd.date_range(start='2020-01-01', periods=36, freq='MS')

# Température simulée : tendance + saisonnalité + bruit
temperature = (
    16                                  # température moyenne
    + 6 * np.sin(2 * np.pi * dates.month / 12)  # saisonnalité annuelle
    + np.random.normal(0, 0.8, size=36)         # bruit
)
# Étape 3 : Création du DataFrame
df = pd.DataFrame({'Date': dates, 'Température (°C)': temperature})
df.set_index('Date', inplace=True)
```

```python
'''Étape 4 : Visualisation de la série temporelle'''
plt.figure(figsize=(10, 4))
plt.plot(df.index, df['Température (°C)'], color='blue', marker='o')
plt.title("Température mensuelle simulée (2020-2022)")
plt.xlabel("Date")
plt.ylabel("Température (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
